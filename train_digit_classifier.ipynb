```
from model_build import model_net
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
import h5py
```

```
print("[INFO] Accessing MNIST data")
((train_X, train_Y),(test_X, test_Y)) = mnist.load_data()

train_X = train_X.reshape((train_X.shape[0],28,28,1))
test_X = test_X.reshape((test_X.shape[0],28,28,1))

train_X = train_X.astype("float32")/255
test_X = test_X.astype("float32")/255

le = LabelBinarizer()
train_Y = le.fit_transform(train_Y)
test_Y = le.transform(test_Y)
```

```
print("[INFO] Compiling model")
init_learning_rate = 1e-3
EPOCHES = 20
mini_batch_size = 128

opt = Adam(lr=init_learning_rate)
model = model_net.build_model(height=28,width=28,depth=1,classes = 10)
model.compile(loss="categorical_crossentropy", metrics = ["accuracy"], optimizer = opt)

model.fit(train_X, train_Y, validation_data = (test_X,test_Y), batch_size = mini_batch_size, epochs = EPOCHES, verbose = 1)
```

```
print("[INFO] Evaluating network")

predict = model.predict(train_X)
print(classification_report(train_Y.argmax(axis=1), predict.argmax(axis=1)))

print("[INFO] Evaluating network")
predictions = model.predict(test_X)
print(classification_report(test_Y.argmax(axis=1),predictions.argmax(axis=1), target_names = [str(x) for x in le.classes_]))
```

```
train_loss,train_acc = model.evaluate(train_X,train_Y)
print('Test accuracy',train_acc)

test_loss,test_acc = model.evaluate(test_X,test_Y)
print('Test accuracy',test_acc)
```
